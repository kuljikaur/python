#### python loops 

# for loop

```python
for i in range(1,10): # this is a range function, which has a strating parameter and a ending parameter, if we just give one parameter,
    #it consider it as ending parameter and gives the starting parameter by itself 

    # also it runs one less from the given limit 
    print(i)
```

```python
for i in range(11):
    print(i)
```

#### break stmt in for loop 

```python
for j in range(11):
    if j == 5:
       break

    print(j)
```

#### continue stmt

###### used to skip an given iteration 

```python
for k in range(11):
    if k == 2:
        continue 

    print(k)
```

#### while loop 

```python
# printing number from 1 to 10 using while loop 
```

```python
p= 1
while(p<10): #1<10, PRINTS, 
    print(p)

    p = p+1 # P =2
```

#### printing numbers from 1 to 10 in reverse order 

```python
for z in range(11, 0, -1): # here satrting value is 11, and ending value is 0 , this -1 is step size, nagative sigs makes numbe decreses 
    print(z)     # next number = current number + step size 
                                  #      11     +  (-1)
                  # next number =       11-1 = 10, 10 + (-1) = 10-1 = 9            
```
