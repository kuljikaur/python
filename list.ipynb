#### finding a largest elemnt in list


for finding largest number in a list, in string, in a number , first consider any number, so that we can compare all the number with that considered number....

```python
lis = [23,45,67,89,12]
largest = lis[0]

for i in lis:
    if i > largest :
        largest = i 

print("the largets no. in the list is : ", largest )
```

#### finding a smaller number in list

```python
lis2 = [2,4,5]
smaller = lis2[0]

for j in lis2:
    if smaller > j :
        smaller = j 

print(smaller)
```
