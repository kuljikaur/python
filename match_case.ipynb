#### switch case or match case 

```python
def switch_stmts(code):
    match code:
        case 200:
            return "ok"

        case 20000:
            return "ok again"

        case 404:
            return "not found"

        case 402:
            return "bad request"

        case _ :
            return "unknown error"

code = int(input("enter any code "))
switch_stmts(code)
```
